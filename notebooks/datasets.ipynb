```python
import pandas as pd
import numpy as np
```

```python
#Load dataset
df = pd.read_csv('heart_disease_uci.csv')
```

```python
print(df.shape)
print(df.isnull().sum())
```

```python
# Drop columns with too many missing values (ca, thal, slope) 
df_cleaned = df.drop(columns=['ca', 'thal', 'slope', 'id', 'dataset'])
```

```python
# Fill remaining missing values with the median (for numeric) or mode (for categorical)
df_cleaned['trestbps'] = df_cleaned['trestbps'].fillna(df_cleaned['trestbps'].median())
df_cleaned['chol'] = df_cleaned['chol'].fillna(df_cleaned['chol'].median())
```
